```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

# Gapminder Dataset

```python
# Example data
gap = px.data.gapminder()
gap.head()
```

```python
gap.shape
```

## Line Plot

```python
# Line Plot
fig = px.line(
    gap,
    x = 'year',
    y = 'lifeExp',
    facet_col = 'continent',
    color = 'country'
    )
fig.show()
```

## Box Plot

```python
#Box Plot
fig = px.box(
    gap,
    x="year",
    y="lifeExp",
    color="continent",
    notched=True)
fig.show()
```

## Scatter Plot

```python
# Scatter Plot
fig = px.scatter(
    gap,
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    animation_frame="year",
    hover_name="country",
    log_x=True,
    size_max=60
    )
fig.update_yaxes(range=[0, 100])
fig.show()
```

## Map

```python
fig = px.choropleth(
    gap,
    locations="iso_alpha",
    color="lifeExp",
    hover_name="country",
    animation_frame="year",
    range_color=[20,80]
)
fig.show()
```

# Iris Dataset

## Scatter Plot w/ Marginal Plots & Trendline

```python
iris = px.data.iris()
fig = px.scatter(
    iris,
    x="sepal_width",
    y="sepal_length",
    color="species",
    marginal_y="violin",
    marginal_x="box",
    trendline="ols"
)
fig.show()
```

## Parallel Coordinates Plot

```python
fig = px.parallel_coordinates(
    iris,
    color="species_id",
    color_continuous_scale=px.colors.diverging.Tealrose,
    color_continuous_midpoint=2
    )
fig.show()
```

## 3D Plot

```python
fig = px.scatter_3d(
    iris,
    x="sepal_width",
    y="sepal_length",
    z="petal_width",
    color="species_id",
    size="species_id",
    hover_name="species_id"
    )
fig.show()
```

## Violin Plot

```python
cars2 = cars[cars['peak_hour'] < 5]
fig = px.violin(
    cars2,
    y="car_hours",
    x="peak_hour",
    color = 'peak_hour',
    box=True,
    points = 'all',
    hover_data=cars.columns
)
fig.show()
```

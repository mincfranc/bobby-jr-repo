```python
projects = !gcloud --format=json projects list | jq -r .[].projectId
projects
```

```python
type(projects)
```

```python
!gcloud --format=yaml projects list
```

```python
for project in projects:
  !gcloud projects describe {project}
  print()
```

# Storage

```python
!gcloud help storage ls
```

```python
for project in projects:
  print(f"== {project}")
  !gcloud storage ls --project={project}
  print()
```

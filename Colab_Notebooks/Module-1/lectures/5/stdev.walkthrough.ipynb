# Summary Stats from Scratch using Python

## Plain Vanilla Python

```python
import random
```

### Creating a list of numbers


```python
size = 1000
values = list( range( 1, size + 1 ) )
random.seed(42)
random.shuffle(values)
values[:5]
```

### Mean

```python
total = sum(values)
f"{total:_}"
```

```python
count = len(values)
count
```

```python
# mean
mean = total/count
mean
```

### Standard Deviation

```python
delta = [ x - mean for x in values ]
delta[:5]
```

```python
delta_squared = [ x**2 for x in delta ]
delta_squared[:5]
```

```python
sum_delta_squared = sum( delta_squared )
sum_delta_squared
```

```python
sample_variance = sum_delta_squared/( count - 1)
sample_variance
```

```python
stdev = sample_variance ** 0.5
stdev
```

## Using numpy in Python

### The fast way

```python
import numpy as np
```

```python
size = 100_000_000
values = np.arange( 1, size + 1 )
np.random.seed(42)
np.random.shuffle(values)
values[:5]
```

```python
values.mean()
```

```python
values.std()
```

### From scratch


```python
total = values.sum()
total
```

```python
count = len( values )
count
```

```python
mean = total / count
mean
```

```python
delta = values - mean
delta[:5]
```

```python
delta_squared = delta**2
delta_squared[:5]
```

```python
sum_delta_squared = delta_squared.sum()
sum_delta_squared
```

```python
sample_variance = sum_delta_squared / ( count - 1 )
sample_variance
```

```python
stdev = sample_variance**0.5
stdev
```

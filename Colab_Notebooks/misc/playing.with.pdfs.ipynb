```python
with pdfplumber.open('comptia-state-of-the-tech-workforce-2024.pdf') as pdf:
    page = pdf.pages[142]
    text = page.extract_text()
    print(text)
```

```python
with pdfplumber.open('comptia-state-of-the-tech-workforce-2024.pdf') as pdf:
    page = pdf.pages[142]
    tables = page.extract_tables()

len(tables)
```

```python
data = []
for line in text.split("\n")[7:58]:
  line_rev = line[::-1]
  fields_rev = line_rev.split(" ",maxsplit=6)
  fields = [ field[::-1].replace(",","").replace("%","") for field in fields_rev ][::-1]
  fields[1:-1]  = [ int(i) for i in fields[1:-1] ]
  fields[-1:]  = [ float(i) for i in fields[-1:] ]
  data += [fields]

data
```

```python
df = pd.DataFrame(data)
df
```
